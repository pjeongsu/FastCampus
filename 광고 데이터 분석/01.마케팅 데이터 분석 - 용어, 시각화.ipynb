### # Domain Knowledge 광고성과지표
| 광고성과지표 | = | 계산식 |
|:---------- |---------- |:---------- |
|CTR(Click Through Rate)|=|클릭수 / 노출수 * 100|
|CPM(Cost Per Mile)|=|광고비용 / 노출수 * 1000|
|CPC(Cost Per Click)|=|광고비용 / 클릭수|
|CPA(Cost Per Action)|=|광고비용 / 구매수|

```python
imp=10000 # 노출수
clk=100 # 클릭수
conv=10 # 구매수
cost=100000 # 광고비용
```

```python
# ctr
ctr = clk/imp*100
```

```python
#ctr 출력
ctr
```

```python
# cpm
cpm = cost/conv*1000
```

```python
#cpm 출력
cpm
```

```python
# cpc
cpc = cost/clk
```

```python
#cpc 출력
cpc
```

```python
# cpa 
cpa = cost/conv
```

```python
#cpa 출력
cpa
```

### # 데이터프레임 시각화

```python
#데이터프레임 변수 생성
dict_data={"철수":[1,2,3,4],"영희":[2,3,4,5],"민수":[3,4,5,6],"수진":[4,5,6,7]}
df=DataFrame(dict_data)
```

```python
df
```

### # 차트 그리기

```python
# 선그래프 
df.plot()
plt.show()

# 데이터프레임의 인덱스가 x축의 범위로 결정됨
# 데이터프레임에서 최소값과 최대값으로 y축의 범위가 결정됨
```

```python
# 막대그래프
df.plot.bar()
plt.show()
```

```python
# 가로막대그래프
df.plot.barh()
plt.show()
```

```python
# 히스토그램
df.plot.hist()
plt.show()
```

```python
# 히스토그램 구간설정
df.plot.hist(bins=range(1, 9, 1)) # x축의 구간 설정
plt.show()
```

### # 차트에 옵션 추가하기

```python
# 기본 막대그래프
df.plot.bar()
plt.show()
```

```python
# 그래프 크기 설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.show()
```

```python
# 제목설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제')
plt.show()
```

```python
# 제목 폰트 크기 설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.show()
```

```python
# x축 이름 설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel')
plt.show()
```

```python
# x축 이름 및 폰트크기 설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.show()
```

```python
# y축 이름 및 폰트 크기 설정 
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.show()
```

```python
# x축 눈금설정

df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize=16, rotation=0) # 설정할 눈금의 위치, 눈금의 이름, 폰트사이즈, 각도
plt.show()
```

```python
# y축 눈금설정

df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize=16, rotation=0) # 설정할 눈금의 위치, 눈금의 이름, 폰트사이즈, 각도
plt.yticks([1, 3, 5, 7], ['첫째', '둘째', '셋째', '넷째'], fontsize=16, rotation=0)
plt.show()
```

```python
# x축 범위설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize=16, rotation=0) # 설정할 눈금의 위치, 눈금의 이름, 폰트사이즈, 각도
plt.yticks([1, 3, 5, 7], ['첫째', '둘째', '셋째', '넷째'], fontsize=16, rotation=0)
plt.xlim([-1, 4])
plt.show()
```

```python
# y축 범위설정
df.plot.bar(figsize=[10, 6]) # 가로, 세로 사이즈
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째', '둘째', '셋째'], fontsize=16, rotation=0) # 설정할 눈금의 위치, 눈금의 이름, 폰트사이즈, 각도
plt.yticks([1, 3, 5, 7], ['첫째', '둘째', '셋째', '넷째'], fontsize=16, rotation=0)
plt.xlim([-1, 4])
plt.ylim([1, 8])
plt.show()
```
